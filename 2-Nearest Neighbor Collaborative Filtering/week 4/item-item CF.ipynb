```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

import warnings
warnings.simplefilter("ignore", UserWarning)
```

## Part 1 - simple ratings:

```python
raw_ratings = pd.read_excel('Assignment 5.xlsx', sheet_name='Ratings').fillna(0)
raw_ratings
```

### Fillind out matrix dataframe based on cosine similarity, derived from user ratings:

```python
matrix_df = pd.read_excel('Assignment 5.xlsx', sheet_name='Matrix', index_col=0)
fill_index, fill_cols = matrix_df.index, matrix_df.columns
matrix_df = pd.DataFrame(cosine_similarity(raw_ratings.values[0:-1,1:-1].T),
                        index = fill_index, columns = fill_cols)
matrix_df
```

### Top five similar movies to Toy story:

```python
matrix_df['1: Toy Story (1995)'].sort_values(ascending=False)[1:6]
```

### Calculating user-item scores:

```python
def calc_ui_score(x, user_name):

    pickeduser_df = raw_ratings[raw_ratings['User'] == user_name]
    m1 = (pickeduser_df != 0).all()
    neigh_list = m1.index[m1.values].tolist()[:-1]
    neigh_list.remove('User')

    w_ij = x.loc[neigh_list].values
    r_uj = pickeduser_df[neigh_list].values[0]
    
    denom = w_ij.sum()


    score = np.dot(w_ij, r_uj) / denom
    return score


scores = matrix_df.apply(calc_ui_score, args=(5277,))
scores.sort_values(ascending = False)[0:5]
```
