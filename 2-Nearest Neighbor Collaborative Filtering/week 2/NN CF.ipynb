```python
import pandas as pd
import numpy as np
```

```python
movie_df = pd.read_csv('movie-row.csv')
corr_df = pd.read_csv('correlations.csv', index_col=0)
user_df = pd.read_csv('user-row.csv', index_col=0)

movie_df.rename({'Unnamed: 0' : 'movie_name'}, axis=1, inplace=True)
```

## Part 1:

```python
def fill_corr(x):
    corr_list = []
    for i in x.index.tolist():
        key = str(x.name)
        corr_list.append(movie_df[key].corr(movie_df[str(i)]))
    return corr_list

corr_df = corr_df.apply(lambda x: fill_corr(x))

corr_df.head()
```

### 4:
Identify the top 5 neighbors (the users with the 5 largest, positive correlations) for users 3867 and 89.

```python
print("Top 5 for 3867: ", corr_df['3867'].sort_values(ascending=False).iloc[1:6].index.tolist())
print("Top 5 for 89: ", corr_df['89'].sort_values(ascending=False).iloc[1:6].index.tolist())
```

### 5:

```python
movie_dft = movie_df.transpose()
movie_dft.fillna(0, inplace=True)


user1 = int(input())
user1_top_neighbors = corr_df[str(user1)].sort_values(ascending=False).iloc[1:6].index.tolist()

def find_best_movies(x):
    numerator = 0
    deno = 0
    for i in user1_top_neighbors:
        if int(x[str(i)]) == 0:
            continue
        deno += corr_df[str(i)].loc[user1]
        numerator += corr_df[str(i)].loc[user1] * float(x[str(i)])
    if deno == 0:
        return 0
    return (x['movie_name'], numerator / deno)

pred_user1 = movie_dft.apply(lambda x: find_best_movies(x))

pred_user1.transpose().sort_values(by=[1], ascending=False)[0:3]
```
