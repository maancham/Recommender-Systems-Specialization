```python
import pandas as pd
import numpy as np
```

```python
from scipy import optimize
```

```python
df = pd.read_excel('Assignment 2.xls', index_col=0)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
feature_name = ['baseball', 'economics', 'politics', 'Europe', 'Asia', 'soccer', 'war',
       'security', 'shopping', 'family']
df.loc[:,['User 1','User 2']].fillna(0,inplace=True)

DFrow = df.loc[['DF']].loc[:,feature_name]
DFrow = DFrow.apply(lambda x: 1/x)

for f in feature_name:
    df.loc['User1', f] = (df[f] * df['User 1']).sum()
    df.loc['User2', f] = (df[f] * df['User 2']).sum()

df['Pred1'] = df.loc[:,feature_name].dot(df.loc[:,feature_name].iloc[-2])
df['Pred2'] = df.loc[:,feature_name].dot(df.loc[:,feature_name].iloc[-1])
```

```python
df.iloc[0:20]['Pred2'].lt(0).sum()
```

### Section 2:

```python
df.loc[:,feature_name] = df.apply(lambda row: row[feature_name] / np.sqrt(row['num-attr']),axis=1)

for f in feature_name:
    df.loc['User1', f] = (df[f] * df['User 1']).sum()
    df.loc['User2', f] = (df[f] * df['User 2']).sum()

df['Pred1'] = df.loc[:,feature_name].dot(df.loc[:,feature_name].iloc[-2])
df['Pred2'] = df.loc[:,feature_name].dot(df.loc[:,feature_name].iloc[-1])

df.iloc[0:20]['Pred1'].sort_values(ascending=False)
```
