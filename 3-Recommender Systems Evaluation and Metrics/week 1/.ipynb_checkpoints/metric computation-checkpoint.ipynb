```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Metric-SpreadSheet-Assignment.xlsx')
df = df.iloc[:, :-3]
```

```python
user_ratings = df[1:11]
header = df.iloc[0]
user_ratings.columns = header.fillna('Users')
user_ratings.set_index('Users', inplace=True)
user_ratings
```

```python
user_preds = df[15:25]
user_preds.columns = header.fillna('Users')
user_preds.set_index('Users', inplace=True)
user_preds
```

### 1: MAE for each user:

```python
mae_user = df[27:37]
mae_user.columns = header.fillna('Users')
mae_user.set_index('Users', inplace=True)
by_user_mae = mae_user.mean(axis=1).sort_values(ascending=False)
by_user_mae
```

### 2: MAE for each movie:

```python
by_movie_mae = mae_user.mean(axis=0).sort_values(ascending=False)
by_movie_mae
```

### 3: Three models of MAE:

```python
print("normal MAE: ", mae_user.sum().sum() / mae_user.count().sum())
print("by-user MAE: ", by_user_mae.mean())
print("by-movie MAE: ", by_movie_mae.mean())
```

### 4: MSE and RMSE for each user:

```python
# mse:
mse_series = np.square(mae_user).sum(axis=1) / mae_user.count(axis=1)

# rmse:
rmse_series = np.sqrt(mse_series)
```

### 5: Overall MSE and RMSE:

```python
print("Overall MSE: ", np.square(mae_user).sum().sum() / np.square(mae_user).count().sum())
print("Overall RMSE: ", np.sqrt(np.square(mae_user).sum().sum() / np.square(mae_user).count().sum()))
```
